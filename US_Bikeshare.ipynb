```python
import os
import pandas as pd

path='/content/drive/MyDrive/US Bikeshare'
```

```python
files =os.listdir(path)
master=pd.DataFrame()

for f in files:
  print(f)

  print('-----')
  df= pd.read_csv(os.path.join(path, f))
  df['city']=f.split('.')[0]
  master = pd.concat([master,df],ignore_index=True)

display(master)
```

```python
day = pd.to_datetime(master['Start Time'])
monthday = day.dt.month
monthday.value_counts()


```

```python
from datetime import datetime

monthlist=[]
start_time = master['Start Time']
for st in start_time:
  date= datetime.strptime(st,"%Y-%m-%d %H:%M:%S")
  month = date.month
  monthlist.append(month)


#datetime.strptime("%Y-%m-%d %H:%M:%S")
```

```python
from collections import Counter
countermonth = Counter(monthlist)
most_common_month = countermonth.most_common(1)

print(countermonth)
print(most_common_month)
```

```python
dayofweeklist=[]
hourofday=[]
for hr in start_time:
  date= datetime.strptime(hr,"%Y-%m-%d %H:%M:%S")
  week = date.strftime('%A')
  dayofweeklist.append(week)
  hour=date.hour
  hourofday.append(hour)
```

```python
weekday=day.dt.dayofweek
weekday
weekday2 = weekday.map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})
weekday3 =weekday2.value_counts()
weekday3
```

```python
hourday=day.dt.hour
hourday.value_counts()
```

```python
counterweekday = Counter(dayofweeklist)
most_weekday = counterweekday.most_common(1)

print(counterweekday)
print(most_weekday)
```

```python
counterhour = Counter(hourofday)
most_hour = counterhour.most_common(1)

print(counterhour)
print(most_hour)
```
